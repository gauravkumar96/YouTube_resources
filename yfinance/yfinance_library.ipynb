```python
import yfinance as yf
import pandas as pd
```

### Getting the OHLC data

```python
# specifying the start and the end date

yf.download("TCS.NS", start="2017-01-01", end="2020-01-01")
```

```python
# specifying only the start date

yf.download("RELIANCE.NS", start = '2011-01-01')
```

```python
# fetching the data for multiple stocks 

yf.download("RELIANCE.NS TCS.NS ONGC.NS", start = '2011-01-01')
```

```python
# fetching intraday data

# 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo

yf.download("RELIANCE.NS", start = '2022-06-01',interval='5m')
```

```python
# fetching 1 minute data for stocks 

yf.download('RELIANCE.NS TCS.NS', start = '2022-07-07',interval='1m')
```
